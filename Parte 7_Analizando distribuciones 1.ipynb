```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Cargar el conjunto de datos
data = pd.read_csv('heart_failure_dataset.csv')

# Filtrar por género (0: mujer, 1: hombre)
female_data = data[data['sex'] == 0]
male_data = data[data['sex'] == 1]

# Obtener la cantidad de anémicos, diabéticos, fumadores y muertos para cada género
female_counts = female_data[['anaemia', 'diabetes', 'smoking', 'DEATH_EVENT']].sum()
male_counts = male_data[['anaemia', 'diabetes', 'smoking', 'DEATH_EVENT']].sum()

# Configurar el tamaño de la figura
plt.figure(figsize=(12, 8))

# Definir el número de barras (variables)
n_bars = len(female_counts.index)

# Definir la anchura de las barras y la separación entre grupos
bar_width = 0.35
bar_positions_male = np.arange(n_bars)
bar_positions_female = bar_positions_male + bar_width

# Crear las barras para hombres a la izquierda
plt.bar(bar_positions_male, male_counts, width=bar_width, color='lightblue', edgecolor='black', label='Hombre')

# Crear las barras para mujeres a la derecha
plt.bar(bar_positions_female, female_counts, width=bar_width, color='lightcoral', edgecolor='black', label='Mujer')

# Configurar el eje x
plt.xticks(bar_positions_male + bar_width / 2, female_counts.index)

plt.title('Distribución de Edades por Género y Condición')
plt.xlabel('Condición')
plt.ylabel('Cantidad')
plt.legend(title='Género')
plt.show()
```
