```python
import pandas as pd
import matplotlib.pyplot as plt

# Paso 1: Cargar el dataframe
df = pd.read_csv('heart_failure_dataset.csv')

# Paso 2: Crear subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Paso 3: Graficar gráficas de torta para cada variable
variables = ['anaemia', 'diabetes', 'smoking', 'DEATH_EVENT']
labels = ['No Anémicos', 'Anémicos'], ['No Diabéticos', 'Diabéticos'], ['No Fumadores', 'Fumadores'], ['Vivos', 'Muertos']

for variable, label, ax in zip(variables, labels, axes.flatten()):
    counts = df[variable].value_counts()
    ax.pie(counts, labels=label, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
    ax.set_title(f'Distribución de {label[1]}')

# Ajustar el diseño de los subplots
plt.tight_layout()

# Paso 4: Mostrar la gráfica
plt.show()
```
